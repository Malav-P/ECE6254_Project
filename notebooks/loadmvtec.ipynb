```python
import torch
from torchvision import transforms
from PIL import Image
```

```python
path_to_image = "../datasets/toothbrush/train/good/000.png"
img = Image.open(path_to_image)
img
```

```python
transform = transforms.Compose([
    transforms.Resize(256),                           # Resize to 256 x 256
    transforms.CenterCrop(224),                        # Crop the center 224x224 region
    transforms.ToTensor(),                             # Convert PIL image to tensor
    transforms.Normalize(mean=[0.485, 0.456, 0.406],  # Normalize the image
                         std=[0.229, 0.224, 0.225])
])

img_tensor = transform(img)
```

```python
img_tensor.shape
```

```python
tensor_image = torch.clamp(img_tensor, 0, 1)  # Ensure the tensor values are between 0 and 1
to_pil = transforms.ToPILImage()
pil_image = to_pil(tensor_image)

pil_image
```
